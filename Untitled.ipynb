```python
import fugashi
import pandas as pd

pd.options.display.max_rows = 20
pd.options.display.min_rows = 20
pd.options.display.max_columns = 30
```

```python
text = """
それは、不思議な夢だった。私は、誰かに追いかけられていた。何かから、必死で逃げていた。不思議だったのは、私が夢の中で自分の子供を連れていたことだ。この子を守らなくては、と子供の手を引き、必死で走っていた。目覚めてからも、しばらくは、つないでいた手の感触が、はっきりと残っていた。
私には子供などいない。結婚すらしていない。大学を卒業し、この四月から働き始めたばかりだった。なのになぜ、あんな夢を見たんだろう。それに何だろう、このせつない気持ちは…。しかし、しばらくは奇妙に思ったものの、朝の忙しさの中で、そんなことは忘れてしまっていた。このアパートには、引っ越してきたばかりだ。慣れない事ばかりで、バタバタの毎日だった。
""".strip()
```

```python
tagger = fugashi.Tagger()
```

```python
attr = ["word", "orth", "lemma", "pos1", "pos2", "cType", "kana", "aType"]

data = pd.concat([pd.DataFrame.from_dict({"word": str(token), **token.feature._asdict()}, orient="index").T
                  for token in tagger(text)]).reset_index(drop=True)
data[attr]
```
